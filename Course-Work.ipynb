```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("slovenia_data.xlsx", sheet_name="Data")
```

```python
selected_columns = [
    'Year', 'GDP (current US$)', 'GDP (constant 2015 US$)', 'GDP growth (annual %)',
    'Exports of goods and services (% of GDP)',
    'Imports of goods and services (% of GDP)',
    'Labor force, total', 'Population, total',
    'Inflation, consumer prices (annual %)'
]
data = data[selected_columns]
```

```python
# Секция 1: ВВП
plt.figure(figsize=(11, 7))
plt.plot(data['Year'], data['GDP (current US$)'], marker='o', label='ВВП (текущие $ США)')
plt.title('Динамика ВВП Словении')
plt.xlabel('Год')
plt.ylabel('ВВП, текущие $ США')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11, 6))
plt.plot(data['Year'], data['GDP (constant 2015 US$)'], marker='o', label='ВВП (константа на 2015 $ США)')
plt.title('Динамика ВВП Словении')
plt.xlabel('Год')
plt.ylabel('ВВП, текущие $ США')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11, 6))
plt.plot(data['Year'], data['GDP growth (annual %)'], marker='o', color='green', label='Темпы роста ВВП')
plt.title('Темпы роста ВВП Словении')
plt.xlabel('Год')
plt.ylabel('Годовой прирост ВВП (%)')
plt.grid()
plt.legend()
plt.show()
```

```python
# Секция 2: Внешняя торговля
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Year', y='Exports of goods and services (% of GDP)', label='Экспорт (% от ВВП)', color='blue')
sns.lineplot(data=data, x='Year', y='Imports of goods and services (% of GDP)', label='Импорт (% от ВВП)', color='red')
plt.title('Экспорт и импорт как доля ВВП')
plt.xlabel('Год')
plt.ylabel('% от ВВП')
plt.legend()
plt.grid()
plt.show()
```

```python
# Секция 3: Инфляция
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Inflation, consumer prices (annual %)'], marker='o', color='orange', label='Инфляция (годовое изменение CPI)')
plt.title('Динамика инфляции в Словении')
plt.xlabel('Год')
plt.ylabel('Инфляция (%)')
plt.grid()
plt.legend()
plt.show()
```

```python
# Секция 4: Трудовые ресурсы
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Labor force, total'], marker='o', color='purple', label='Численность рабочей силы')
plt.title('Динамика численности рабочей силы в Словении')
plt.xlabel('Год')
plt.ylabel('Численность рабочей силы')
plt.grid()
plt.legend()
plt.show()
```

```python
# Секция 5: Демография
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Population, total'], marker='o', color='brown', label='Население')
plt.title('Динамика численности населения Словении')
plt.xlabel('Год')
plt.ylabel('Численность населения')
plt.grid()
plt.legend()
plt.show()
```

```python
# Итоговая корреляция
plt.figure(figsize=(10, 6))
corr_data = data.drop(columns=['Year']).corr()
sns.heatmap(corr_data, annot=True, cmap='coolwarm')
plt.title('Корреляция экономических показателей')
plt.show()
```
